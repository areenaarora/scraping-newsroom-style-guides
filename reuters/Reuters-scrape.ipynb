```python
html_file = open("reuters-styleguide.html", "r")
  
# Reading the file
rawhtml = html_file.read()
```

```python
from bs4 import BeautifulSoup

text = BeautifulSoup(rawhtml, 'html.parser')
```

```python
all_entries = text.find_all(['h2', 'p'])[177:]
```

```python
all_entries
```

```python
reuters_list_of_dicts = []

word = ''
meaning = ''

for each_entry in all_entries:
#   print(each_entry)
# h2 = word, p = its description
    if each_entry.name == 'h2':
        reuters_list_of_dicts.append({'word_reuters': word, 'meaning_reuters': meaning})
        word = each_entry.text
        meaning = ''
#         print(each_entry.text)
    else:
        meaning = meaning + each_entry.text
reuters_list_of_dicts.append({'word_reuters': word, 'meaning_reuters': meaning})
# print(reuters_list_of_dicts)
```

```python
import pandas as pd

df_reuters = pd.DataFrame(reuters_list_of_dicts)
pd.set_option('display.max_colwidth', None)
```
