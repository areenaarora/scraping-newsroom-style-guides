```python
import requests
from bs4 import BeautifulSoup
```

```python
bbc_url = "https://www.bbc.co.uk/newsstyleguide/all/"
raw_html = requests.get(bbc_url).content
```

```python
soup_bbc = BeautifulSoup(raw_html, "html.parser")
```

```python
all_text = soup_bbc.find_all('p')[1:]
```

```python
# find word using strong tag
# find meaning that follows the strong tag and store it in the word
```

```python
bbc_list = []

for each_entry in all_text:
    if each_entry.contents[0].name == 'strong' and len(each_entry.contents) == 1:
        print(each_entry.text)
```

```python
bbc_list = []

word = ""
word_meaning = ""

for each_entry in all_text:
    if each_entry.contents[0].name == 'strong' and len(each_entry.contents) == 1:
        bbc_list.append({"word_bbc": word, "meaning_bbc": word_meaning.replace(u"\xa0", u" ")})
        word = each_entry.text
        word_meaning = ""
    else:
        word_meaning = word_meaning + each_entry.text
bbc_list.append({"word_bbc": word, "meaning_bbc": word_meaning})
# final append to include Zelensky because first append was doing so only if there was
# an entry after word. There wasn't anything after the last word, so it didn't take that one
```

```python
import pandas as pd

df_bbc = pd.DataFrame(bbc_list)
pd.set_option('display.max_colwidth', None)
```
