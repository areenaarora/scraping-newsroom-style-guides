```python
import requests
from bs4 import BeautifulSoup
```

```python
buzzfeed_url = "https://www.buzzfeed.com/emmyf/buzzfeed-style-guide"
raw_html = requests.get(buzzfeed_url).content
```

```python
soup_buzzfeed = BeautifulSoup(raw_html, "html.parser")
```

```python
each_meaning = soup_buzzfeed.find_all("p")
```

```python
buzzfeed_list_of_dicts_try = []

for definitions in each_meaning[21:561]:
    print(definitions.contents)
    word = definitions.contents[0].text
    description = definitions.contents[1:]
    word_meaning = ""
    for each_line_part in description:
        word_meaning = word_meaning + str(each_line_part.text).strip()
    buzzfeed_list_of_dicts_try.append({'word_buzzfeed': word, 'meaning_buzzfeed': word_meaning})
```

```python
import pandas as pd
df_BuzzFeed = pd.DataFrame(buzzfeed_list_of_dicts_try)
pd.set_option('display.max_colwidth', None)
```
